```python
import random

def generate_test_case(truck_dimensions, truck_weight_capacity, num_items):
    truck_length, truck_width, truck_height = truck_dimensions
    truck_volume = truck_length * truck_width * truck_height
    max_item_volume = truck_volume // num_items
    max_item_weight = truck_weight_capacity // num_items
    
    items = []
    total_volume = 0
    total_weight = 0
    
    for i in range(1, num_items + 1):
        while True:
            length = random.randint(1, truck_length)
            width = random.randint(1, truck_width)
            height = random.randint(1, truck_height)
            weight = random.randint(1, max_item_weight)
            item_volume = length * width * height

            if (total_volume + item_volume <= truck_volume) and (total_weight + weight <= truck_weight_capacity):
                total_volume += item_volume
                total_weight += weight
                items.append((i, (length, width, height), weight))
                break

    return {"truck": (truck_dimensions, truck_weight_capacity), "items": items}

def write_test_case_to_file(test_case, file_name):
    with open(file_name, 'w') as f:
        truck = test_case["truck"]
        items = test_case["items"]
        
        f.write(f"Truck: {truck}\n")
        f.write("Items:\n")
        for item in items:
            f.write(f"{item}\n")

# Example usage
random.seed(42)  # For reproducibility

# Define truck specifications and number of items
truck_spec_1 = ((15, 10, 8), 1000)
num_items_1 = 10

truck_spec_2 = ((20, 15, 12), 2000)
num_items_2 = 15

# Generate test cases
test_case_1 = generate_test_case(*truck_spec_1, num_items_1)
test_case_2 = generate_test_case(*truck_spec_2, num_items_2)

# Write test cases to text files
write_test_case_to_file(test_case_1, 'test_case_1.txt')
write_test_case_to_file(test_case_2, 'test_case_2.txt')

print("Test cases written to 'test_case_1.txt' and 'test_case_2.txt'")
```

```python
import csv

def get_column_values(file_name, column_name):
    values = []
    with open(file_name, mode='r') as file:
        csv_reader = csv.DictReader(file)
        for row in csv_reader:
            if column_name in row:
                values.append(row[column_name])
            else:
                raise ValueError(f"Column '{column_name}' not found in the CSV file.")
            if len(values) == 30:
                break;
    return values

# Example usage
file_name = 'city_16_divisions.csv'
column_name = 'division_code'

# Call the function and print the values
load_locs = get_column_values(file_name, column_name)

print(load_locs)
```
